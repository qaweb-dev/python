**Фрактали** — це геометричні об'єкти, які мають властивість самоподібності, тобто їх структура повторюється на різних масштабах. Вони можуть бути створені математичними алгоритмами або виникати природно, як форми гір, хмар, рослин чи берегових ліній. Вперше термін "фрактал" ввів математик Бенуа Мандельброт у 1975 році, описуючи складні форми, що не піддаються класичній геометрії. Фрактали широко застосовуються в комп'ютерній графіці, фізиці, біології, а також для аналізу фінансових ринків та інших складних систем.

# Наведемо приклад коду Python побудови фрактального дерева

```python
import matplotlib.pyplot as plt
import numpy as np

def fractal_tree(ax, x, y, length, angle, depth):
    if depth == 0:
        return
    x_end = x + length * np.cos(np.radians(angle))
    y_end = y + length * np.sin(np.radians(angle))
    ax.plot([x, x_end], [y, y_end], color='brown', lw=1)
    fractal_tree(ax, x_end, y_end, length*0.7, angle+30, depth-1)
    fractal_tree(ax, x_end, y_end, length*0.7, angle-30, depth-1)

fig, ax = plt.subplots()
ax.set_aspect('equal', adjustable='box')
ax.axis('off')
fractal_tree(ax, 0, 0, 1, 90, 7)
plt.show()
```
