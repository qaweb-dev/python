**Що таке програмування?**

Програмування — це процес створення інструкцій для комп'ютера, щоб він міг виконувати певні завдання. Ці інструкції називаються кодом, і вони пишуться спеціальними мовами програмування.

**Чому варто вчитися програмуванню?**

* **Творчість:** Програмування дозволяє створювати щось нове, від простих програм до складних застосунків.
* **Розв'язання задач:** Програмування допомагає розв'язувати різноманітні задачі, від автоматизації рутинних процесів до аналізу великих обсягів даних.
* **Затребуваність:** Програмісти затребувані в багатьох галузях, що відкриває широкі можливості для кар'єрного зростання.
* **Розвиток мислення:** Програмування розвиває логічне та аналітичне мислення.

**З чого почати?**

Для початку, давайте познайомимося з однією з найпопулярніших і найпростіших у вивченні мов програмування — Python.

**Python: ваша перша мова програмування**

Python — це універсальна мова програмування, яка використовується для веб-розробки, аналізу даних, штучного інтелекту та багато чого іншого. Вона відома своїм простим і зрозумілим синтаксисом, що робить її ідеальною для початківців.

**Ваша перша програма на Python: "Hello, World!"**

Традиційно, перша програма, яку пишуть новачки, виводить на екран фразу "Hello, World!". Ось як це зробити на Python:

```
print("Hello, World!")
```

**Як запустити програму:**

1.  **Встановіть Python:** Якщо у вас ще не встановлено Python, завантажте та встановіть його з офіційного сайту python.org.
2.  **Створіть файл:** Відкрийте текстовий редактор (наприклад, Блокнот) і скопіюйте туди код. Збережіть файл із розширенням .py (наприклад, hello.py).
3.  **Запустіть програму:** Відкрийте командний рядок або термінал, перейдіть до папки, де збережено файл, і виконайте команду `python hello.py`.

Ви побачите, як на екрані з'явиться фраза "Hello, World!". Вітаю, ви написали свою першу програму!

**Додаткові приклади коду для початківців:**

1.  **Змінні та типи даних:**

```
name = "Іван"  # Змінна типу рядок (string)
age = 30  # Змінна типу ціле число (integer)
height = 1.75  # Змінна типу число з плаваючою комою (float)

print("Мене звати", name)
print("Мій вік:", age)
print("Мій зріст:", height)
```

Пояснення:

* Змінні використовуються для зберігання даних.
* Тип даних визначає, які значення може зберігати змінна.
* Функція `print()` виводить значення на екран.

2.  **Арифметичні операції:**

```
x = 10
y = 5

sum = x + y  # Додавання
difference = x - y  # Віднімання
product = x * y  # Множення
quotient = x / y  # Ділення

print("Сума:", sum)
print("Різниця:", difference)
print("Добуток:", product)
print("Частка:", quotient)
```

Пояснення:

* Python підтримує основні арифметичні операції.

3.  **Умовні оператори:**

```
age = 20

if age >= 18:
    print("Ви повнолітній")
else:
    print("Ви неповнолітній")
```

Пояснення:

* Умовні оператори дозволяють виконувати код залежно від умови.
* `if` перевіряє умову, `else` виконується, якщо умова хибна.
