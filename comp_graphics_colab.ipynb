# Приклади обробки графіки

### Імпортуємо потрібні бібліотеки

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

### Отримуємо та відтворюємо зображення

Далі: перший рядок отримує зображення, другий - змінює схум кольорів, третій - створює холст (полотно), настурний - виодить зображення у холст, останній рядок - показує зображення. Зображення було завантажене в Google Colab для поточного прикладу.

```python
image = cv2.imread('/adapter-bluetooth.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

fig = plt.figure()
plt.imshow(image)
plt.show()
```

Для скорочення коду в наступних прикладах створюємо власну функцію, яка перетворює схему BRG в RGB та показує зображення.

```python
def show_img(image):
    image_to_show = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    fig = plt.figure()
    plt.imshow(image_to_show)
    plt.show()
```

Покажемо як взяти зображення з мережі та виконати зміну (resize) його розмірів.

```python
import urllib

url = 'https://qaweb.dev/files/icons/logo-gaweb.jpg'
resp = urllib.request.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype="uint8")
image = cv2.imdecode(image, cv2.IMREAD_COLOR)
image = cv2.resize(image,(160, 40))
show_img(image)
```

Виконаємо перетворення кольорового зображенні у сіре, а також виконаємо його бінаризацію (два кольори - білий і чорний) та виділення контуру зображення.

```python
import urllib

url = 'https://qaweb.dev/files/icons/logo-gaweb.jpg'
resp = urllib.request.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype="uint8")
image = cv2.imdecode(image, cv2.IMREAD_COLOR)
show_img(image)

gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
show_img(gray_image)

ret,thresh_binary = cv2.threshold(gray_image,127,255,cv2.THRESH_BINARY)
show_img(thresh_binary)

thresh_mean = cv2.adaptiveThreshold(gray_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
show_img(thresh_mean)
```
